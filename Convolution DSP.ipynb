```python
def con(x,h):
    conv = []
    h = h[::-1]
    xpad = [0]*(len(h)-1)+x+[0]*(len(h)-1)
    for i in range(0,len(x)+len(h)-1):
            y = 0
            for j in range(0, len(h)):
                y += h[j]*xpad[i + j]
            conv.append(y)
    return conv
con1 = con(x,h)
print(con1)
```
